```python
def limpiar_lista_dataframe(lista):
  '''
  Recibe la lista con DF y devuelve la misma lista con datos limpios
  '''
  lista_limpia = []
  for df in lista:
    lista_limpia.append(limpiar_df(df.df))
  return lista_limpia
```

```python
def limpiar_df(df, espacios=range(0,2), cabecera=1):
  '''
  Devuelve el DF limpio
  '''
  nuevo_df = df.drop(espacios,axis=0).reset_index(drop=True)
  
  tabla_final.drop(range(0,2),axis=0).reset_index(drop=True)
  return nuevo_df
```
